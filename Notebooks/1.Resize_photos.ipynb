# Resize Photos
This notebook contains the code needed to resize the photos of the motorcycles.
Running this notebook will create a new folder with the images rescaled.
This first step is needed in order to make the computations feasible in a local machine.

## Libraries

```python
# To access files
import os
import cv2 # OpenCV-Python
# To show images
%matplotlib inline
import matplotlib.pyplot as plt
# To time the performance
import time
```

## Example of resizing

### Examination

```python
# Original path where all the subfolders are
data_original_path = "../../Circuito_Almeria_171021/Iniciados/Data/"
```

```python
folder_number = os.listdir(data_original_path)[0]
# Show the name of the folder
folder_number
```

```python
# Example of the name of a photo
os.listdir(data_original_path + os.listdir(data_original_path)[0])[1]
```

```python
# Reading with openCV an example of an image
moto_example = cv2.imread(data_original_path + folder_number + "/" + os.listdir(data_original_path + os.listdir(data_original_path)[0])[5])
```

```python
# Show the original shape
moto_example.shape
```

```python
# Show the image
plt.imshow(cv2.cvtColor(moto_example, cv2.COLOR_BGR2RGB))
plt.show()
```

One can see, on the axis, the number of pixel in each edge.

### Resize

```python
# Set the percentage of the resolution desired
scale_percent = 5
# THE NEW DIMENSIONS ARE COMPUTED
width = int(moto_example.shape[1] * scale_percent / 100)
height = int(moto_example.shape[0] * scale_percent / 100)
dim = (width, height)

# Resizing the image
resized = cv2.resize(moto_example, dim, interpolation = cv2.INTER_AREA)
```

```python
# Print the new shape
resized.shape
```

```python
# The resized image is shown
plt.imshow(cv2.cvtColor(resized, cv2.COLOR_BGR2RGB))
plt.show()
```

## Resizing all images

### To a 5 %

```python
start_time = time.time()
# Path where the data will be saved
data_path = "../Circuito_Almeria_171021/Iniciados/Data_resized/"
# Iterate through the folders of the original path
for folder in os.listdir(data_original_path):
    # Print folder as log
    print(data_original_path + folder)
    # Create folder in the specified directory
    os.mkdir(data_path + folder)
    # Iterate through the original images of the folder
    for file in os.listdir(data_original_path + folder):
        # Read the image
        img = cv2.imread(data_original_path + folder + "/" + file)
        # Resize the image
        resized = cv2.resize(img, dim, interpolation = cv2.INTER_AREA)
        # Write the resized image
        cv2.imwrite(data_path + folder + "/" + file, resized)
# Print total time elapsed
print('Elapsed time: %.6fs' % (time.time() - start_time))
```
